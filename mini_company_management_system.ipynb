```python
import datetime

class Employee:
  employee_count={}
  def __init__(self,name,joining_date,work_experience=0,weekly_work_hour=40):
    self.name=name
    self.joining_date=joining_date
    self.work_experience=work_experience
    self.weekly_work_hour=weekly_work_hour

    if 40 <= weekly_work_hour <= 60:
      self.weekly_work_hour = weekly_work_hour
    elif weekly_work_hour < 40:
      print(f"{self.name} can not work for {weekly_work_hour} hours.")
      self.weekly_work_hour = 40
    elif weekly_work_hour > 60:
      print(f"{self.name} can not work more then 60 hours.")
      self.weekly_work_hour = 40

  @classmethod
  def showDetails(cls):
    print("Company workforce:")

    if cls.employee_count:
      total_employee = sum(cls.employee_count.values())
      print(f"Total Employee/s : {total_employee}")
      if 'programmer' in cls.employee_count:
        print(f"Total Programmer Employee/s: : {cls.employee_count['programmer']}")
      if 'HR employee' in cls.employee_count:
        print(f"Total HR Employee/s : {cls.employee_count['HR employee']}")
    else:
      print("Total Employee/s : 0")



class Programmer(Employee):
  designation_list=[ 'Junior Software Engineer', 'Software Engineer', 'Senior Software Engineer', 'Technical Lead' ]
  def __init__(self,name,joining_date,work_experience=0,weekly_work_hour=40):
    super().__init__(name,joining_date,work_experience,weekly_work_hour)
    self.id = self.createProgrammerID()
    self.salary = 0
    if 'programmer' not in Employee.employee_count:
      Employee.employee_count['programmer'] = 1
    else:
      Employee.employee_count['programmer'] += 1

    self.designation = ""
    if 0 <= self.work_experience <= 3:
      self.designation = Programmer.designation_list[0]
    elif 3 < self.work_experience <= 5:
      self.designation = Programmer.designation_list[1]
    elif 5 < self.work_experience <= 8:
      self.designation = Programmer.designation_list[2]
    elif self.work_experience > 8:
      self.designation = Programmer.designation_list[3]

  def createProgrammerID(self):
    year = self.joining_date[2:4]
    month = self.joining_date[5:7]
    day = self.joining_date[8:10]
    count = 0
    for value in self.employee_count.values():
      count += value
    return f"P-{year}{month}{day}-{count}"

  def calculateSalary(self):
    if self.designation == Programmer.designation_list[0]:
      salary = 30000
    elif self.designation == Programmer.designation_list[1]:
      salary = 45000
    elif self.designation == Programmer.designation_list[2]:
      salary = 70000
    else:
      salary = 120000

    today = datetime.date.today()
    year = today.year
    join_year = int(self.joining_date[0:4])
    final_year = year - join_year

    if self.salary == 0:
      self.salary = salary
    for i in range(1,final_year+1):
      s = (0.15*1)*self.salary
      self.salary += s

  def calculateOvertime(self):
    overtime = self.weekly_work_hour - 40
    if overtime == 0:
      print(f"{self.name} will not get overtime.")
    else:
      overtime_money = (overtime*500)*4
      self.salary += overtime_money
      print(f"{self.name} will get BDT {overtime_money} overtime.")

  def showProgrammerDetails(self):
    year = self.joining_date[2:4]
    month = self.joining_date[5:7]
    day = self.joining_date[8:10]
    join = f"{day}-{month}-{year}"
    print("Programmer Employee:")
    print(f"Name: {self.name}")
    print(f"ID: {self.id}")
    print(f"Joining Date: {join}")
    print(f"Designation: {self.designation}")
    print(f"Salary: BDT {self.salary}")



class HR(Employee):
  def __init__(self , name , joining_date , work_experience , weekly_work_hour = 40):
    super().__init__(name , joining_date , work_experience , weekly_work_hour)
    if 'HR employee' not in Employee.employee_count:
      Employee.employee_count['HR employee'] = 1
    else:
      Employee.employee_count['HR employee'] += 1
    self.id = self.createHREmployeeID()

  def createHREmployeeID(self):
    year = self.joining_date[2:4]
    month = self.joining_date[5:7]
    day = self.joining_date[8:10]
    count = 0
    for value in self.employee_count.values():
      count += value
    return f"HR-{year}{month}{day}-{count}"

  def showHREmployeeDeatails(self):
    year = self.joining_date[2:4]
    month = self.joining_date[5:7]
    day = self.joining_date[8:10]
    join = f"{day}-{month}-{year}"
    print("HR Employee:")
    print(f"Name:: {self.name}")
    print(f"ID: {self.id}")
    print(f"Joining Date: {join}")




class InternProgrammer(Employee):
  intern_count = 0
  def __init__(self, name, joining_date, intern_type = "Unpaid"):
    super().__init__(name, joining_date)
    InternProgrammer.intern_count += 1
    self.temp_id = "Temp_"+str(self.intern_count)
    self.intern_type = intern_type
    self.work_experience = 0
    self.weekly_work_hour = 40
    self.status = ""

    today = datetime.date.today()
    current_year = int(today.year)
    current_month = int(today.month)
    current_day = int(today.day)
    intern_join_year = int(self.joining_date[0:4])
    intern_join_month=int(self.joining_date[5:7])
    intern_join_day=int(self.joining_date[8:10])

    if current_month - int(intern_join_month) >= 4:
      self.status = 'Eligible for promotion'
    else:
      self.status = 'Not Eligible for promotion'
    self.joining_date = f"{intern_join_day}-{intern_join_month}-{intern_join_year}"
    self.id = 'Temp_' + str(InternProgrammer.intern_count)


  def showInternDetails(self):
    print("Intern (Programmer):")
    print(f"Name: {self.name}")
    print(f"ID: {self.id}")
    print(f"Joining date: {self.joining_date}")
    print(f"Type: {self.intern_type}")
    print(f"Status : {self.status}")

  def promoteToProgrammer(self):
    if self.status:
      today = datetime.date.today()
      curr_date = str(today.day)
      curr_month = str(today.month)
      curr_year = str(today.year)
      if len(curr_month) == 1:
        curr_month = '0' + curr_month
      if len(curr_date) == 1:
        curr_date = '0' + curr_date
      print(f'{self.name } is promoted!')
      programmer = Programmer(self.name , f"{curr_year}-{curr_month}-{curr_date}" , 0 , 40)
      return programmer
    else:
      print(f"{self.name} can not be promoted.")




Employee.showDetails()
print("=========1=========")
richard = Programmer("Richard Hendricks", "2021-06-08", 4, 48)
richard.calculateSalary()
print("=========2=========")
richard.showProgrammerDetails()
print("=========3=========")
richard.calculateOvertime()
print("=========4=========")
richard.showProgrammerDetails()
print("=========5=========")
monica = HR("Monica Hall", "2022-07-06", 2, 40)
print("=========6=========")
monica.showHREmployeeDeatails()
print("=========7=========")
Employee.showDetails()
print("=========8=========")
gilfoyle = Programmer("Bertram Gilfoyle", "2020-03-02", 6, 35)
gilfoyle.calculateSalary()
print("=========9=========")
gilfoyle.calculateOvertime()
print("=========10=========")
gilfoyle.showProgrammerDetails()
print("=========11=========")
gavin = Programmer("Gavin Belson", "2016-12-20", 9)
gavin.calculateSalary()
gavin.calculateOvertime()
gavin.showProgrammerDetails()
print("=========12=========")
yang = InternProgrammer("Jian Yang", "2023-01-01")
yang.showInternDetails()
print("=========13=========")
jared = InternProgrammer("Jared Dunn", "2023-06-05", "Paid")
jared.showInternDetails()
print("=========14=========")
jared = jared.promoteToProgrammer()
print("=========15=========")
Employee.showDetails()
print("=========16=========")
yang = yang.promoteToProgrammer()
yang.calculateSalary()
yang.showProgrammerDetails()
print("=========17=========")
Employee.showDetails()
```
